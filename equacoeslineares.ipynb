```python
import sympy as sp
from IPython.display import display
x, y, z = sp.symbols('x y z')
print("resolva o sistema de equações de 3 variáveis")
print("x+y+z=6")
print("2x-y+z=3")
print("x+2y-z=2")
print("solução:")
sol = sp.solve([
    sp.Eq(x + y + z, 6),
    sp.Eq(2*x - y + z, 3),
    sp.Eq(x + 2*y - z, 2)
], (x, y, z))

display(sol)
```
